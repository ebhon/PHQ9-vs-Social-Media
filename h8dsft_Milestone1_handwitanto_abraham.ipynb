## Replace Value

```python
#Mengganti Value pada kolom jawaban kuisioner dengan angkat 0-3 sesuai dengan penghitungan PHQ9
phq9 = {
    'Not at all':'0',
    'Several days':'1',
    'Half of days':'2',
    'Nearly everyday':'3'
}
df['little_interest_doing_things'] = df['little_interest_doing_things'].replace(phq9)
df['depressed_or_hopeless'] = df['depressed_or_hopeless'].replace(phq9)
df['trouble_sleeping'] = df['trouble_sleeping'].replace(phq9)
df['feeling_tired'] = df['feeling_tired'].replace(phq9)
df['poor_apetite'] = df['poor_apetite'].replace(phq9)
df['trouble_concentrating'] = df['trouble_concentrating'].replace(phq9)
df['moving_speaking_slow_vice_versa'] = df['moving_speaking_slow_vice_versa'].replace(phq9)
df['thoughts_bad'] = df['thoughts_bad'].replace(phq9)
df['feeling_bad'] = df['feeling_bad'].replace(phq9)

#Membuat variabel dan mengubah variabel yang berisi rentang waktu agar dapat di analisa
df['own_sm'] = df['sm_duration_owned']
owned_sm = {
    'Less than 2-year' : '2',
    '2-5 years': '5',
    '5-10 years': '10',
    'More than 10 years' : '11'
}
df['own_sm'] = df['own_sm'].replace(owned_sm)

#Membuat variabel dan mengubah value yang berisi rentang waktu agar dapat di analisa
df['spend_on_sm'] = df['daily_spend_on_sm']
daily_spend = {
    'Less than 1 hour':'1',
    '1-3 hours': '3',
    '3-5 hours' : '5',
    'More than 5 hours' : '6'
}
df['spend_on_sm'] = df['spend_on_sm'].replace(daily_spend)

#mengubah tipe data tabel terkait menjadi Numeric untuk keperluan analisa data
df[['little_interest_doing_things', 'depressed_or_hopeless','trouble_sleeping', 'feeling_tired',
    'poor_apetite','trouble_concentrating','moving_speaking_slow_vice_versa','thoughts_bad',
    'feeling_bad',]] = df[['little_interest_doing_things', 'depressed_or_hopeless','trouble_sleeping', 'feeling_tired',
    'poor_apetite','trouble_concentrating','moving_speaking_slow_vice_versa','thoughts_bad',
    'feeling_bad']].apply(pd.to_numeric)

#mengubah tipe data timestamp menjadi Datetime
df['timestamp'] = df['timestamp'].apply(pd.to_datetime)

#Membuat kolom Score yang berisi jumlah penghitungan PHQ9
df['score'] = df['little_interest_doing_things']+df['depressed_or_hopeless']+df['trouble_sleeping']+df['feeling_tired']+df['poor_apetite']+df['feeling_bad']+df['trouble_concentrating']+df['moving_speaking_slow_vice_versa']+df['thoughts_bad']
```

```python
#Membuat klasifikasi score sesuai perhitungan PHQ9
def klasifikasi_score(score):
    if score >= 0 and score <= 4:
        return 'Normal'
    elif score >= 5 and score <= 9:
        return 'Depresi Ringan'
    elif score >=10 and score <= 14:
        return 'Depresi Sedang'
    elif score >=15 and score <=19:
        return 'Depresi Berat'
    else:
        return 'Depresi Sangat Berat'
df['depression_level'] = df['score'].apply(klasifikasi_score)

#Membuat klasifikasi pendapatan bulanan menjadi angkat agar dapat di analisa
def klasifikasi_pendapatan(monthly_income):
    if monthly_income == 'No income':
        return 0
    elif monthly_income == '10,000-40,000 Tk':
        return 1
    elif monthly_income == '40,000-70,000 Tk':
        return 2
    else:
        return 3
df['income_id'] = df['monthly_income'].apply(klasifikasi_pendapatan)

#membuat klasifikasi pendidikan menjadi level angka untuk analisa data
def klasifikasi_pendidikan(education):
    if education == 'None':
        return 0
    elif education == 'Primary level':
        return 1
    elif education == 'Secondary Level':
        return 2
    elif education == 'Higher secondary level':
        return 3
    elif education == 'Graduate Level':
        return 4
    elif education == 'Graduate or above':
        return 5
    else:
        return 6
df['education_id'] = df['education'].apply(klasifikasi_pendidikan)

def klasifikasi_jumlah_teman(friends_num_sm):
    if friends_num_sm == 'Less than 500':
        return 0
    elif friends_num_sm == '500-2000':
        return 1
    elif friends_num_sm == '2000-4000':
        return 2   
    else:
        return 3
df['friends_num_id'] = df['friends_num_sm'].apply(klasifikasi_jumlah_teman)

def klasifikasi_real_friend(personal_friend_sm):
    if personal_friend_sm == 'Few of them':
        return 0
    elif personal_friend_sm == 'Many of them':
        return 1
    elif personal_friend_sm == 'Most of them':
        return 2   
    else:
        return 3
df['real_friends_id'] = df['personal_friend_sm'].apply(klasifikasi_real_friend)

def klasifikasi_post(post_frequency):
    if post_frequency == 'Less than 1 per day':
        return 0
    elif post_frequency == ' 1-2 per day':
        return 1
    elif post_frequency == '3-5 per day':
        return 2   
    else:
        return 3
df['post_freq_id'] = df['post_frequency'].apply(klasifikasi_post)

def klasifikasi_profession(profession):
    if profession == 'Unemployed including students':
        return 1
    elif profession == 'Housewife':
        return 2
    elif profession == 'Daily labor/driver/housemaid':
        return 3
    elif profession == 'Private job':
        return 4
    elif profession == 'Self-employed/Entrepreneur/Business':
        return 5 
    elif profession == 'Govt. job':
        return 6      
    else:
        return 7
df['profession_id'] = df['profession'].apply(klasifikasi_profession)

def klasifikasi_gender(gender):
    if gender == 'Female':
        return 1    
    else:
        return 2
df['gender_id'] = df['gender'].apply(klasifikasi_gender)

def klasifikasi_status(marital_status):
    if marital_status == 'Unmarried':
        return 1
    elif marital_status == 'Widow':
        return 2
    elif marital_status == 'Divorced':
        return 3  
    else:
        return 4
df['marital_id'] = df['marital_status'].apply(klasifikasi_status)

def klasifikasi_residence(residence):
    if residence == 'Urban Area':
        return 1    
    else:
        return 2
df['residence_id'] = df['residence'].apply(klasifikasi_residence)
```

```python
#Penggunaan Z-Score untuk menghitung batas bawah dan batas atas, karena persebaran data "spend_on_sm" adalah normal
avg = df['spend_on_sm'].mean()
std = df['spend_on_sm'].std()

up_bound = avg + 3*std
low_bound = avg - 3*std

print (f"Upper Boundary = {up_bound}")
print (f"Lower Boundary = {low_bound}")
```

#### Outlier

#### Korelasi durasi penggunaan sosial media dan tingkat depresi

```python
#Menganalisa korelasi durasi penggunaan sosial media dengan tingkat depresi menggunakan kendall correlation. 
corr_tau, pval_k = stats.kendalltau(df['score'], df['spend_on_sm'])
print(f"tau_correlation: {corr_tau:.2f}, p_value: {pval_k}")
```

##### Insight : Korelasi durasi menggunakan sosial media dengan tingkat depresi
- tau_correlation: 0.14 : Terdapat korelasi pada keduanya, namun kecil atau bisa di asumsikan ada faktor lain yang ikut mempengaruhi tingkat depresi responden<br>
- p_value: 2.2941001996972573e-07 : Nilai p-value jauh dibawah 0.05, artinya terdapat bukti untuk melawan H0, sehingga benar terdapat korelasi antara durasi menggunakan sosial media dengan tingkat depresi responden.

Dikarenakan sudah terdapat bukti bahwa terdapat korelasi antara durasi menggunakan sosial media dengan tingkat depresi responden, maka perlu dilakukan analisa lebih lanjut untuk mengetahui variabel apa yang memiliki hubungan dengan kenaikan durasi penggunaan sosial media.
